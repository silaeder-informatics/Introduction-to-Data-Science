```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_features=2, random_state=42, cluster_std=2, n_samples=500)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(X, y)
```

```python
from sklearn.externals.six import StringIO  
from sklearn.tree import export_graphviz
from IPython.display import Image  
import pydotplus

#       graphviz,    .
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, impurity=False, feature_names=["X1", "X2"])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
import numpy as np

xx, yy = np.meshgrid(np.linspace(X[:,0].min(), X[:,0].max(), 20),
                     np.linspace(X[:,1].min(), X[:,1].max(), 20))

points = np.stack([xx, yy], axis=-1).reshape(-1, 2)

plt.contourf(xx, yy, tree.predict(points).reshape(20,20))
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(X, y)
y_predicted = tree.predict(X)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y, y_predicted))
```
